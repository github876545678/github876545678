Bilet 4 Sual 1

```python
from collections import Counter

# İstifadəçidən mətn alırıq
text = input("Mətni daxil edin: ")

# Sözlərin sayını tapırıq
words = text.split()
word_count = len(words)

# Ən uzun sözün uzunluğunu tapırıq
longest_word_length = max(len(word) for word in words)

# Ən çox istifadə edilən hərfi tapırıq
letters = [char for char in text if char.isalpha()]  # Yalnız hərfləri seçirik
most_common_letter = Counter(letters).most_common(1)[0]

# Nəticələri çap edirik
print(f"Sözlərin sayı: {word_count}")
print(f"Ən uzun sözün uzunluğu: {longest_word_length}")
print(f"Ən çox istifadə edilən hərf: '{most_common_letter[0]}', İstifadə sayı: {most_common_letter[1]}")
```

```python


# Salary sütunundakı boş dəyərləri median ilə doldururuq
median_salary = df['Salary'].median()
df['Salary'].fillna(median_salary, inplace=True)

# Dəyişiklikləri yoxlayırıq
print("Datasetdəki boş dəyərlərdən sonra Salary sütunu:")
print(df['Salary'])
```

```python

import seaborn as sns
import matplotlib.pyplot as plt


# Seaborn ilə scatter plot yaratmaq
plt.figure(figsize=(10, 6))  # Şəkil ölçüsünü təyin edirik
sns.scatterplot(x='Age', y='Salary', data=df, color='blue')

# Qrafikin başlığını və ox etiketlərini təyin edirik
plt.title('Age və Salary Arasındakı Əlaqə', fontsize=14)
plt.xlabel('Yaş', fontsize=12)
plt.ylabel('Maaş', fontsize=12)

# Qrafiki göstəririk
plt.show()
```

```python
department_counts = df['Department'].value_counts()

# Bar chart yaratmaq
plt.figure(figsize=(10, 6))  # Şəkil ölçüsünü təyin edirik
department_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Qrafikin başlığını və ox etiketlərini təyin edirik
plt.title('Department üzrə Paylanma', fontsize=14)
plt.xlabel('Department', fontsize=12)
plt.ylabel('Say', fontsize=12)

# Qrafiki göstəririk
plt.show()
```

```python
# Müştəridən məlumatları alırıq
income = float(input("Aylıq gəlirinizi daxil edin (AZN): "))
expenses = float(input("Aylıq ümumi xərclərinizi daxil edin (AZN): "))
debt = float(input("Ümumi borcunuzu daxil edin (AZN): "))

# Aylıq gəlir ilə aylıq xərclər arasındakı fərqi tapırıq
income_expenses_difference = income - expenses

# Fərqi yüzdə 20 ilə müqayisə edirik
if income_expenses_difference / income < 0.2:
    print("Fərq 20%-dən azdır. Təəssüf ki, kredit verə bilməyəcəyik.")
else:
    # Əgər fərq 20%-dən çoxdursa, borcun gəlirin 40%-dən çox olmaması şərti ilə kredit verə bilərik
    if debt / income < 0.4:
        print("Kredit verə bilərsiniz. Borcunuz gəlirinizin 40%-dən azdır.")
    else:
        print("Kredit verə bilməyəcəyik. Borcunuz gəlirinizin 40%-dən çoxdur.")
```
