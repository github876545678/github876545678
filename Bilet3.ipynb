```python
import pandas as pd
import matplotlib.pyplot as plt

positions = df['POSITION'].astype(str) # Convert 'positions' to string type

ages = df['BIRTH_DATE']

# Nöqtə qrafiki qururuq
plt.figure(figsize=(10, 6))
plt.scatter(positions, ages, alpha=0.7, color='blue')

# Qrafikə başlıq və ox adlarını əlavə edirik
plt.title('POSITION və BIRTH_DATE arasındakı əlaqə', fontsize=14)
plt.xlabel('POSITION', fontsize=12)
plt.ylabel('BIRTH_DATE', fontsize=12)
plt.xticks(rotation=45)  # Kateqorial dəyərləri daha oxunaqlı etmək üçün
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Qrafiki göstəririk
plt.tight_layout()
plt.show()
```

```python
# İşçinin məlumatlarını istifadəçidən alırıq
name = input("İşçinin adını daxil edin: ")
hourly_wage = float(input("Saatlıq maaşı daxil edin: "))
hours_worked = float(input("İşlədiyi saatların sayını daxil edin: "))

# Əsas əmək haqqını hesablayırıq
if hours_worked > 40:
    # Artıq işlənən saatları tapırıq
    overtime_hours = hours_worked - 40

    # Əsas və əlavə maaşı hesablayırıq
    regular_pay = 40 * hourly_wage
    overtime_pay = overtime_hours * (hourly_wage * 1.5)
    total_pay = regular_pay + overtime_pay
else:
    # Əgər 40 saatdan az işləyibsə
    total_pay = hours_worked * hourly_wage

# Nəticəni çap edirik
print(f"{name} adlı işçinin ümumi əmək haqqı: {total_pay:.2f} AZN")
```

```python
import numpy as np

# Verilmiş massiv
data = np.array([[[1, 2, 3, 4, 5], [11, 21, 31, 41, 51]],
                 [[11, 12, 13, 14, 15], [51, 52, 53, 54, 5]]])

# Sonuncu alt massivi seçirik
last_subarray = data[-1, -1]

# Yeni massivdə ən böyük və ən kiçik elementləri tapırıq
max_element = np.max(last_subarray)
min_element = np.min(last_subarray)

# Hər bir alt massiv üzrə elementlərin cəmini hesablayırıq
sums = np.sum(data, axis=2)

# Nəticələri çap edirik
print("Sonuncu alt massiv:", last_subarray)
print("Ən böyük element:", max_element)
print("Ən kiçik element:", min_element)
print("Hər bir alt massiv üzrə elementlərin cəmi:")
print(sums)
```
