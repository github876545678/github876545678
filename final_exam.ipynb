```python
import re

df['PHONE_NUMBERS_ONLY'] = df['TELEPHONE'].apply(lambda x: ''.join(re.findall(r'\d', str(x))))

# Yeni DataFrame ilə nəticəni göstəririk
print(df)
```

```python

from datetime import datetime

df['BIRTH_DATE'] = pd.to_datetime(df['BIRTH_DATE'])

# Cari tarixi alırıq
current_date = datetime.now()

# Yaşı hesablayırıq və yeni AGE sütunu əlavə edirik
df['AGE'] = df['BIRTH_DATE'].apply(lambda x: current_date.year - x.year - ((current_date.month, current_date.day) < (x.month, x.day)))

# Yaşların orta, maksimum və minimum qiymətlərini hesablamaq
average_age = df['AGE'].mean()
max_age = df['AGE'].max()
min_age = df['AGE'].min()

# Nəticələri göstəririk
print(f"Orta yaş: {average_age}")
print(f"Max yaş: {max_age}")
print(f"Min yaş: {min_age}")
```

```python
most_common_position = df['POSITION'].mode()[0]  # Ən çox rast gəlinən dəyər
least_common_position = df['POSITION'].value_counts().idxmin()  # Ən az rast gəlinən dəyər

# Telefon nömrəsi olmayan şəxslərin faizini hesablamaq
no_phone_number_percentage = (df['TELEPHONE'].isnull().sum() / len(df)) * 100

# Nəticələri göstəririk
print(f"Ən çox rast gəlinən POSITION dəyəri: {most_common_position}")
print(f"Ən az rast gəlinən POSITION dəyəri: {least_common_position}")
print(f"Telefon nömrəsi olmayan şəxslərin faizi: {no_phone_number_percentage:.2f}%")
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

df['POSITION'] = df['POSITION'].astype('category').cat.codes

# Scatter qrafikini yaratmaq
plt.figure(figsize=(10,6))
sns.scatterplot(x='POSITION', y='AGE', data=df)

# Başlıqlar və etiketlər əlavə edirik
plt.title('POSITION və AGE Arasındakı Əlaqə')
plt.xlabel('POSITION')
plt.ylabel('AGE')

# Qrafiki göstəririk
plt.show()
```

```python
bins = range(0, 101, 10)  # 0-dan 100-ə qədər, 10-lu addımlarla
labels = [f'{i}-{i+9}' for i in bins[:-1]]  # Qrup etiketlərini yaradırıq

df['Age Group'] = pd.cut(df['AGE'], bins=bins, labels=labels, right=False)

# Hər bir yaş qrupundakı şəxslərin sayını tapırıq
age_group_counts = df['Age Group'].value_counts()

# Pie chart ilə göstəririk
plt.figure(figsize=(8,8))
age_group_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='Set3')

# Başlıq əlavə edirik
plt.title('Yaş Qruplarına Görə Şəxslərin Sayı')

# Qrafiki göstəririk
plt.show()
```

```python
import seaborn as sns

grouped_data = df.groupby(['AREA_ACTIVITY', 'POSITION']).size().unstack(fill_value=0)

# İstilik xəritəsini (heatmap) yaradırıq
plt.figure(figsize=(10, 8))
sns.heatmap(grouped_data, annot=True, cmap='YlGnBu', fmt='d')

# Başlıq əlavə edirik
plt.title('AREA_ACTIVITY və POSITION üzrə Şəxslərin Sayı')

# Qrafiki göstəririk
plt.show()
```

```python
import re

# Parola yoxlama funksiyası
def check_password_strength(password):
    # Şərtləri yoxlamaq üçün hər birini müvafiq regex ilə yoxlayırıq
    if (len(password) >= 8 and  # Parolanın uzunluğu ən azı 8 simvol olmalıdır
        re.search(r'[A-Z]', password) and  # Böyük hərf olmalıdır
        re.search(r'[a-z]', password) and  # Kiçik hərf olmalıdır
        re.search(r'[0-9]', password) and  # Rəqəm olmalıdır
        re.search(r'[@#$%^&+=]', password)):  # Xüsusi simvol olmalıdır
        return "Güclü parola"
    else:
        return "Zəif parola"

# İstifadəçidən parola alırıq
password = input("Parolanızı daxil edin: ")

# Parolanın güclülüyünü yoxlayırıq və nəticəni çap edirik
result = check_password_strength(password)
print(result)
```
