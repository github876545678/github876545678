```python
import librosa
import matplotlib.pyplot as plt
```

```python
audio_df='/content/mixkit-people-indoors-ambience-2981.wav'
audio_df
```

```python
audio_data, sampling_rate= librosa.load(audio_df)
```

```python
import librosa.display as display
import numpy as np
```

```python
y,sr = librosa.load(audio_df)
```

## **1.MFCC Xususiyyetleri**


```python


mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
plt.figure(figsize=(10,4))
librosa.display.specshow(mfcc,x_axis='time')
plt.colorbar()
plt.title("MFCC Xüsusiyyətləri")
plt.show()
```

2.ZCR Xususiyyetleri

```python
zcr= librosa.feature.zero_crossing_rate(y)
plt.figure(figsize=(10,4))
plt.plot(zcr[0],color='r')
plt.title("Zero Crossing Rate(ZCR)")
plt.show()
```

3.Spektral Roll-Off Xususiyyeti

```python
spectral_rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr, roll_percent=0.85)
plt.figure(figsize=(10,4))
plt.semilogy(spectral_rolloff.T, label='Spectral Roll-Off')
plt.ylabel('Hz')
plt.xticks([])
plt.xlim([0, spectral_rolloff.shape[-1]])
plt.title("Spektral Roll-Off Xüsusiyyəti")
plt.show()
```

4.Chroma Xususiyyetleri

```python
chroma = librosa.feature.chroma_stft(y=y, sr=sr)
plt.figure(figsize=(10,4))
librosa.display.specshow(chroma, x_axis='time', y_axis='chroma', cmap='coolwarm')
plt.colorbar()
plt.title('Chroma Xüsusiyyətləri')
plt.show()
```

5.RMS Enerji Xususiyyetleri

```python
rms=librosa.feature.rms(y=y)
plt.figure(figsize=(10,4))
plt.plot(rms[0], label='RMS Enerji', color='b')
plt.title("RMS Enerji Xüsusiyyəti")
plt.show()
```
