```python
import numpy as np

scores = np.array([[[1, 2, 3, 4, 5], [11, 21, 31, 41, 51]],
                   [[11, 12, 13, 14, 15], [51, 52, 53, 54, 5]]])

reversed_scores = np.flip(scores)

print("Əsas massiv:")
print(scores)
print("\nTərs massiv:")
print(reversed_scores)
```

```python
import numpy as np

scores = np.array([[[1, 2, 3, 4, 5], [11, 21, 31, 41, 51]]])

even_numbers = scores % 2 == 0

even_count = np.sum(even_numbers)

print("Cüt ədədlərin sayı:", even_count)
```

```python
import numpy as np

scores = np.array([[[1, 2, 3, 4, 5], [11, 21, 31, 41, 51]],
                   [[11, 12, 13, 14, 15], [51, 52, 53, 54, 5]]])

max_elements = np.max(scores, axis=(1, 2))

sum_max_elements = np.sum(max_elements)

print("Hər bir alt massivdəki ən böyük elementlər:", max_elements)
print("Onların cəmi:", sum_max_elements)
```

```python
text = input("Mətnı daxil edin: ")


def shift_text(text):
    shifted_text = ''
    for char in text:
        if char.isalpha():

            if char.islower():
                shifted_text += chr(((ord(char) - ord('a') + 1) % 26) + ord('a'))

            elif char.isupper():
                shifted_text += chr(((ord(char) - ord('A') + 1) % 26) + ord('A'))
        else:
            shifted_text += char
    return shifted_text

shifted_text = shift_text(text)
print("Şifrələnmiş mətn:", shifted_text)
```
